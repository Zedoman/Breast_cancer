# Breast Cancer Prediction

```python
#loading the dataset through pandas Library
import pandas as pd
df= pd.read_csv('breast-cancer (1).csv')
df.head()
```

```python
#how many rows and cols are there
df.shape
```

```python
#cols name
df.columns
```

# Data Cleaning

```python
#to drop the unwanted cols
df.drop(columns= ['Unnamed: 32','id'], inplace= True)
df.head()
```

# Label Encoder

```python
#doing the label encoder for the col diagnosis
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
df.iloc[:,0]=lb.fit_transform(df.iloc[:,0])
df.head()
```

```python
#feature and targets are being separated
from sklearn.model_selection import train_test_split
x=df.iloc[:,1:-1] #for feature
y=df.iloc[:,0]#for target
x
```

```python
y
```

```python
#doing training,testing 
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

# Data Preprocessing By Norm and SC

# Normalization

```python
#the Values are rescaled so that they end up ranging between 0 to 1 also called minmax scaler
from sklearn.preprocessing import MinMaxScaler
mms=MinMaxScaler()
x_train_norm=mms.fit_transform(x_train)
x_test_norm=mms.fit_transform(x_test)
```

```python
#x_train data being normalized into 0 to 1
x_train_norm
```

```python
#x_test data being normalized into 0 to 1
x_test_norm
```

# Standard Scaling

```python
#here the values are centered around the mean with a unit standard deviation
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
#The fit method is calculating the mean and variance of each of the features present in our data. 
#The transform method is transforming all the features using the respective mean and variance
x_train_sc=sc.fit_transform(x_train)
x_test_sc=sc.fit_transform(x_test)
```

```python
#x_train data being standardized
x_train_sc
```

```python
#x_test data being standardized
x_test_sc
```

# Predictive model

# SVM

```python
# It can handle both classification and regression on linear and non-linear data.
from sklearn import svm
model=svm.SVC()
model.fit(x_train,y_train)
prediction=model.predict(x_test)
prediction
```

```python
y_test
```

```python
count_miss=(y_test!=prediction).sum()
```

```python
count_miss
```

```python
#SVM Accuracy
from sklearn import metrics
accuracy=metrics.accuracy_score(y_test,prediction)
accuracy*100
```

# Decision Tree 

```python
# Decision Trees are excellent tools for helping us to choose between several courses of action.
from sklearn.model_selection import GridSearchCV
from sklearn import metrics
from sklearn import tree
```

```python
#decision tree accuracy
dtree=tree.DecisionTreeClassifier(criterion='entropy',max_depth=10,max_features=5)
dtree.fit(x_train,y_train)
predict=dtree.predict(x_test)
acc=metrics.accuracy_score(y_test,predict)
acc*100
```

# Random Forest

```python
from sklearn.model_selection import GridSearchCV
from sklearn import tree
from sklearn import ensemble
forest_params = [{'max_depth': list(range(10, 15))}]
clf = GridSearchCV(dtree, forest_params, cv = 10, scoring='accuracy')
clf.fit(x_train, y_train)
```

```python
print(clf.best_params_)
```

```python
print(clf.best_score_)
```

```python
#random forest accuracy
rf=ensemble.RandomForestClassifier(n_estimators=10,max_depth=10,criterion='entropy',max_features=5)
rf.fit(x_train,y_train)
pred2=rf.predict(x_test)
accu=metrics.accuracy_score(y_test,pred2)
accu*100
```

# Classification report

# From the above predictive modes Random Forest gives the best accuracy report so we use it for classification report

```python
c_m=metrics.confusion_matrix(y_test,pred2)
```

```python
c_m
```

```python
c_r=metrics.classification_report(y_test,pred2)
print(c_r)
```
